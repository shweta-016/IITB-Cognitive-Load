```python
import pandas as pd
import matplotlib.pyplot as plt

eeg = pd.read_csv("C:/Users/Admin/Desktop/cognitive load/Data//EEG.csv")
gsr = pd.read_csv("C:/Users/Admin/Desktop/cognitive load/Data//GSR.csv")
```

```python
eeg['Delta_avg'] = eeg[['Delta_TP9','Delta_AF7','Delta_AF8','Delta_TP10']].mean(axis=1)
eeg['Theta_avg'] = eeg[['Theta_TP9','Theta_AF7','Theta_AF8','Theta_TP10']].mean(axis=1)
eeg['Alpha_avg'] = eeg[['Alpha_TP9','Alpha_AF7','Alpha_AF8','Alpha_TP10']].mean(axis=1)
```

```python
plt.figure(figsize=(10,5))
plt.plot(eeg['Delta_avg'], label='Delta')
plt.plot(eeg['Theta_avg'], label='Theta')
plt.plot(eeg['Alpha_avg'], label='Alpha')
plt.legend()
plt.title("EEG Band Averages Over Time")
plt.xlabel("Samples")
plt.ylabel("Amplitude")
plt.show()
```

```python
plt.figure(figsize=(10,4))
plt.plot(gsr['GSR Conductance CAL'], color='green')
plt.title("GSR Conductance Over Time")
plt.xlabel("Samples")
plt.ylabel("GSR Value")
plt.show()
```
